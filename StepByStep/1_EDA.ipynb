```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath = 'https://docs.google.com/spreadsheets/d/1_B0reSRNs5_Se-5FPUtUB6AmjT511Ey2e6BjzrD4nY4/export?format=csv&gid=0'
df = pd.read_csv(filepath)
```

```python
quant_cols = ['Rent', 'Deposit', 'Meterage' ]
cat_cols = ['DoB', 'Area', 'District', 'Elevator', 'Parking', 'Storage', 'Balcony' ,  'NoB' , 'FloorNum']
```

```python
sns.histplot(df['Meterage'])
plt.show()
```

```python
for col in quant_cols:
    fig = plt.figure(figsize=(10,8))
    sns.histplot(x=df[col] , kde=True)
    plt.title(f'distribution for {col}')
    plt.show()
```

```python
for col in cat_cols:
    fig = plt.figure(figsize = (12,8))
    sns.countplot(data = df , x = col)
    
```

```python
for col in quant_cols:
    fig = plt.figure(figsize=(8,8))
    sns.boxplot(x=df[col])
    plt.title(f'boxplot for {col}')
    plt.show()
```

```python
for cat in cat_cols:
    for quant in quant_cols:
        fig = plt.figure(figsize = (8,10))
        sns.barplot(x = cat , y = quant , data=df )
        plt.title(f'average {quant} by {cat} ')
```

```python
for cat in cat_cols:
    for quant in quant_cols:
        fig = plt.figure(figsize = (8,10))
        sns.boxplot(x = cat , y = quant , data=df )
        plt.title(f'average {quant} by {cat} ')
```

```python
for i in range(len(quant_cols)):
    for j in range(i + 1 , len(quant_cols)):
        quant1 = quant_cols[i]
        quant2 = quant_cols[j]
        fig = plt.figure(figsize=(4,2))
        sns.scatterplot(x=quant1 , y = quant2 , data=df)

```

```python
sns.pairplot(df[quant_cols] , diag_kind = 'kde')
```

```python
sns.pairplot(df , diag_kind = 'kde')
```

```python
corr_mat = df[quant_cols].corr()
fig = plt.figure(figsize=(4,2))
sns.heatmap(corr_mat , annot = True , cmap='coolwarm' , square = True , vmin = 1 , vmax = 1 )
plt.show()
```

```python
corr_mat = df.drop(['Area','link'] , axis=1).corr()
fig = plt.figure(figsize=(10,8))
sns.heatmap(corr_mat , annot = True , cmap='coolwarm' , square = True  )
plt.show()
```
