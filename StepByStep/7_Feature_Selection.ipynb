```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('dataProcessed')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0' , axis = 1)
```

```python
df.head()
```

```python
X = df.drop(['nRent','Rent','Deposit'] , axis=1)
y = df['nRent']
```

```python
model = LinearRegression()

sfs = SequentialFeatureSelector(model, 
                                k_features='best',  # tedad featurehayy ke mikhaim entekhab konim , ageh ro best bzarim khodesh e
                                # entekhab mikoneh
                                forward=True, # ageh True bashe forewardeh ageh False basheh Backward
                                floating=False, # ageh True bashe yek jor stepwise ro anjam mideh
                                scoring='neg_mean_squared_error', # az manfi MSE estefadeh mikone
                                cv=5) 

                                # bray inkeh az overfitting jologiri konim az cross validation estefadeh mikonim
                                # cross validation miad data maro be 5 fold mokhtalaf taghsim mikoneh va 4 tay on ro train mikoneh 
                                # va yeki az on fold haro bray testing be kar migireh va inkaro ta akhar anjam mideh va bhtrin 
                                # natijeh ro byan mikoneh ba variationhay mokhtalef
sfs.fit(X, y)
print('Selected features:', sfs.k_feature_idx_)
```
